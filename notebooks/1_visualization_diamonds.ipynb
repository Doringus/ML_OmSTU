```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%config InlineBackend.figure_format = 'svg'
```

Загрузим датасет

```python
diamonds_data = pd.read_csv("../data/diamonds/diamonds.csv", index_col = 0)
```

Посмотрим как распределены алмазы по огранке

```python
diamonds_data["cut"].value_counts().plot(kind = 'pie')
```

Видно, что в датасете преобладают алмазы с идеальной, премиумной и очень хорошей огранкой

Построим графики зависимости стоимости от огранки, цвета и чистоты камня

```python
sns.barplot(x = 'cut', y = 'price', data = diamonds_data)
```

```python
sns.barplot(x = 'color', y = 'price', data = diamonds_data)
```

```python
sns.barplot(x = 'clarity', y = 'price', data = diamonds_data)
```

Посмотрим, как выглядит зависимость стоимости от веса алмаза

```python
sns.lineplot(x = 'carat', y = 'price', data = diamonds_data)
```

Была получена нелинейная зависимость. Значит вес алмаза не самый информативный признак.

Построим зависимость стоимости от чистоты для каждого вида огранки

```python
sns.catplot(data=diamonds_data, x='clarity', hue='cut', y='price', kind='point', aspect=2)
```

Видно, что дороже всего алмазы с чистотой SI2

Посмотрим на корреляцию признаков

```python
sns.heatmap(diamonds_data.corr(), annot=True)
```

Видно, что цена зависит от веса, а также ширины, высоты и глубины камня
