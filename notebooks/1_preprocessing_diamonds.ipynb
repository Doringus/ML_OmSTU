```python
import pandas as pd
```

```python
diamonds_data = pd.read_csv("../data/diamonds/diamonds.csv", index_col = 0)
diamonds_data.head()
```

```python
diamonds_data.describe()
```

1. Проверим датасет на наличие пустых ячеек:

```python
diamonds_data.isnull().sum()
```

Пустых признаков в датасете нет

2. В датасете есть алмазы с параметрами x, y, z равными 0. Исключим их из набора.

```python
diamonds_data = diamonds_data.drop(diamonds_data[diamonds_data["x"] == 0].index)
diamonds_data = diamonds_data.drop(diamonds_data[diamonds_data["y"] == 0].index)
diamonds_data = diamonds_data.drop(diamonds_data[diamonds_data["z"] == 0].index)
diamonds_data.describe()
```

3. В датасете присутствуют три категориальных признака, избавимся от них при помощи one-hot кодирования. Предварительно скопируем данные, чтобы не изменить оригинальный датасет.

```python
diamonds_ml = diamonds_data.copy()
```

```python
diamonds_ml = pd.get_dummies(diamonds_data)
diamonds_ml.head()
```

4. Сохраним обработанный датасет

```python
diamonds_ml.to_csv("../data/diamonds/diamonds_preprocessed.csv", sep = ',')
```
